```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy


x = np.linspace(-10, 10, 100)
a1 = 3
a2 = 5
b = 1

y1 = cauchy.pdf(x, a1, b)
y2 = cauchy.pdf(x, a2, b)

plt.plot(x, y1 / (y1 + y2), label='P(w1 | x)')
plt.plot(x, y2 / (y1 + y2), label='P(w2 | x)')

plt.xlabel('x')
plt.ylabel('probability')
plt.title('Posterior Probabilities')

plt.legend()
plt.grid()
plt.savefig('../assets/Q1_posterior.png')
```
